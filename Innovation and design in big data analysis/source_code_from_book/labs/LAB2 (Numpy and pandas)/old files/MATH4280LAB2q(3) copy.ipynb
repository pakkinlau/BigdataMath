## How to find all the local minima (or troughs) in a 1d array?
Write a function to find all the troughs in a 1D numpy array `a`. Troughs are points surrounded by larger values on both sides.

(Hence elements at both ends will not be considered as troughs in any case.)

1st Example input:

```
a = np.array([1, 3, 7, 1, 2, 6, 0, 1])
```

1st Example output:

```
 array([3, 6])
```

where, 3 and 6 are the positions of peak values 1 and 0.

2nd Example input:

```
np.random.seed(100)
a = np.random.randint(10, size=100)
```

2nd Example output:

```
array([ 2,  5,  7, 13, 16, 19, 21, 23, 26, 32, 37, 41, 45, 47, 49, 51, 54,
       56, 58, 60, 62, 66, 68, 75, 77, 79, 82, 84, 88, 90], dtype=int64)
```


````python
## answer 1

def find_troughs(a):
    # input parameter: a is our numpy list.
    # output: returning index of elements which condition met. 
    indices = []
    
    i=1
    
    print(f"\nLength of the vector is: {len(a)}")
    
    # While numpy arrays are zero-indexed, [0,9] is the boundary of the full index.
    # The first and last element won't be troughs. So our loop will run in the range of [1,8]. 
    
    # testing condition:
    # compare the actual value from the index i: a[i], and compare toe neighbors a[i-1], a[i+1]
    # condition: left elt. and right elt larger value 
    # return position of the elt which the condition holds true
    
    # while loop: 
    # Start: we set i = 1 to skip the first element.
    # End: we set the last while loop to run when i = 8,
    # Total: we test 8 elements for 'whether it is a trough' condition.
    
    # So in the for-loop in the code, len(a) - 2 = 8, 
    # So, at the end the code would run the loop in the range of [1,8], which fits our initial setting. 
    
    def is_trough(a,i):
        if (a[i] < a[i-1] and a[i] < a[i+1]):
            return True
        else:
            return False
    
    while i <= len(a)-2:
        
        # printing the current running element for debugging purpose.
        # print(f"current running elt{i}...")
        
        if is_trough(a,i):
            indices.append(i)
        i = i+1
    return indices

def testing_package(a):
    print(f"The original set of random number is: {a}")
    trough_indices = find_troughs(a)
    return a[trough_indices].astype('int64')

print("1st example input (predefined set of values):")
# 1st exampel input:
a = np.array([1, 3, 7, 1, 2, 6, 0, 1])

print("Example output:")
testing_package(a)


````

````python
print("2nd example input (random 100 numbers):")
# 2nd example input:
# With seed = 100, it set a fixed set of random numbers in the global state of numpy.
np.random.seed(100)
b = np.random.randint(10, size=100)

print("Example output:")
testing_package(b)
````

````python
## answer 2c

def swap_index_of_columns(dataframe,):
    all_col_index = dataframe.columns #.tolist()
    sorted_col_index = all_col_index.sort_values()
    reverse_sorted_index = sorted_col_index[::-1]
    return reverse_sorted_index

def swap_columns(dataframe):
    new_column_index = swap_index_of_columns(dataframe)
    new_df = pd.DataFrame(dataframe, columns=new_column_index)
    return new_df

# test 1
a = df.columns
print(f"the column names are: {a}")
new_df = swap_columns(df)
new_df

````

````python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(1,100, 40).reshape(10, -1), columns=list('pqrs'), index=list('abcdefghij'))


lis = df.index.tolist() 

"""
# print(lis): ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
"""

# Iterate over rows and print each row
coor_dict = {}
for index, row in df.iterrows(): 
    # 1. df.iterrows() returns a generator. 2. it splits each row into 2 parts. index and row
    coor_dict[index] = row


from math import sqrt

def calculate_distance(current_point,current_coordinate,each_other_point, each_other_coordinate):
    """_summary_

    Args:
        current_point_coordinate (pd.Series): ...
        other_point_coordinate (dict): ...

    Returns:
        _type_: _description_
    """

    # Now, current_point_coordinate and point are pd.Series.
    
    a = current_coordinate['p'] - each_other_coordinate['p']
    b = current_coordinate['q'] - each_other_coordinate['q']
    c = current_coordinate['r'] - each_other_coordinate['r']
    d = current_coordinate['s'] - each_other_coordinate['s']
    dist = sqrt(a**2 + b**2 + c**2 + d**2)

    return dist            


def find_min_values(data):
    min_values = {}
    for key, inner_dict in data.items():
        min_key = min(inner_dict, key=inner_dict.get)
        min_value = inner_dict[min_key]
        min_values[key] = {min_key: min_value}
    return min_values

def book_keeping_distance(dict_of_info):
    
    list_of_points: list[str] = dict_of_info.keys() # list of points: ['a','b',...]
    
    book_keeping_dict = {} # expect to contain - {point_N : nearest_point, min_distance}
    
    # choosing the subject points 
    for point in list_of_points:
        
        # chopping out 2 parts of data from the dataframe
        temp_dict = dict_of_info.copy()
        
        # prepare 2 sets of coordinates
        current_point_coor_pack = temp_dict.pop(point) 
        
        # print(f" current_point_coor_pack: {current_point_coor_pack}")
        
        # Remaining pairs 
        other_points_coordinate =temp_dict
        
        # this dict tries to keep the shortest distance point and its distance.
        min_distance = np.inf # key is the target point, and its distance
        min_dist_point = ""
        
        min_record = {}
        
        # for-loop run totally len(list_of_points) = around 10 times.
        for each_other_point, each_other_coordinate in other_points_coordinate.items(): # 'point' and 'current_point_coordinate' are pd.Series. 
            
            current_point: str = current_point_coor_pack.name
            current_coordinate = current_point_coor_pack
            
            dist = calculate_distance(current_point,current_coordinate,each_other_point, each_other_coordinate)
            
            min_distance = dist
            min_dist_point = each_other_point

            min_record[min_dist_point] = min_distance

        book_keeping_dict[point] = (min_record)
    return book_keeping_dict
        
shortest_distances = find_min_values(book_keeping_distance(coor_dict))


# Convert the dictionary to a DataFrame
df2 = pd.DataFrame([(outer_key, inner_key, value) for outer_key, inner_dict in shortest_distances.items() for inner_key, value in inner_dict.items()],
                   columns=['outer_key', 'nearest_row', 'dist'])

# Merge the two DataFrames based on common keys
result = pd.merge(df, df2, left_index=True, right_on='outer_key', how='left')

# Set the 'outer_key' as the index and drop the 'outer_key' column
result.set_index('outer_key', inplace=True)
result.index.name = None

print(result)

# Treat every point as a space
````
