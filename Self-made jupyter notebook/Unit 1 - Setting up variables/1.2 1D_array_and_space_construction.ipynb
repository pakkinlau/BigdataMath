## Plot a sphere

Certainly! The code you provided generates the surface coordinates of a sphere in 3D space. The mathematical equations for the surface coordinates of a sphere in terms of spherical coordinates (ρ, θ, φ) are as follows:
- $x = \rho \sin(\phi) \cos(\theta)$
- $y = \rho \sin(\phi) \sin(\theta)$
- $z = \rho \cos(\phi)$

In these equations:
- $x$, $y$, and $z$ represent the Cartesian coordinates of the points on the sphere's surface.
- $\rho$ (rho) represents the radial distance from the origin to the point on the sphere's surface.
- $\phi$ (phi) represents the polar angle, ranging from 0 to $\pi$, which measures the angle down from the positive z-axis.
- $\theta$ (theta) represents the azimuthal angle, ranging from $-\pi$ to $\pi$, which measures the angle in the xy-plane from the positive x-axis.

```python
fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')

# `np.linspace(start, stop, num)` generates an array of `num` evenly spaced values over the interval `[start, stop]`.
# `linspace` ensures that the values of `u` and `v` are uniformly sampled. 
# , which returns a set of evenly spaced numbers / points over a specified interval.
# the set of points are stored as a numpy array.
u = np.linspace(-np.pi, np.pi, 100)
v = np.linspace(0, np.pi, 100)

# the shape of the sphere
# `np.outer(a, b)` computes the outer product of two vectors.
# since u and v is already a numpy array, it can be treated as a vector.

# `np.ones(np.size(u))`: this part creates an array of ones with the same size as `u`.
# `np.ones()` represents \rho = 1. 

x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the surface
surf1 = ax1.plot_surface(x, y, z, cmap='jet',alpha=0.6,facecolors=plt.cm.jet(z),linewidth=0.5,rcount=30,ccount=30)

# Set edge color of the shape to be black
surf1.set_edgecolor('k')

# Set the axis for the chart 
ax1.set_xlim3d(-2, 2)
ax1.set_ylim3d(-2, 2)
ax1.set_zlim3d(-2, 2)

```

In the code:
- $u$ corresponds to $\theta$ in the range from $-\pi$ to $\pi$.
- $v$ corresponds to $\phi$ in the range from 0 to $\pi$.
- $x$ is computed using $\cos(u) \cdot \sin(v)$.
- $y$ is computed using $\sin(u) \cdot \sin(v)$.
- $z$ is computed using $\cos(v)$.

These equations map points (u, v) from the parameter space to points (x, y, z) on the surface of a unit sphere.
