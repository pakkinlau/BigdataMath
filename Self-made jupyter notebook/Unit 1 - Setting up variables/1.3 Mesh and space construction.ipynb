```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
import scipy.interpolate
from matplotlib import rcParams
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

# Set font size for plots
rcParams.update({'font.size': 18})

# Define grid spacing and create 2D grid of x and y values
h = 0.5
x = np.arange(-6, 6 + h, h)
y = np.arange(-6, 6 + h, h)
X, Y = np.meshgrid(x, y)

# Define three 2D functions F0, F, and F2 based on X and Y coordinates
# (But F is not used in the visualization, just for making F2 to be more complicated)
F0 = 1.5 - 1.0 * np.exp(-0.03 * (3 * np.power(X, 2) + np.power(Y, 2)))
F = 1.5 - 1.6 * np.exp(-0.05 * (3 * np.power(X, 2) + np.power(Y + 3, 2)))
F2 = F + (0.5 - 1.0 * np.exp(-0.1 * (3 * np.power(X - 3, 2) + np.power(Y - 3, 2))))

# Calculate partial derivatives of F0 and F2 with respect to x and y using gradient function
dF0y, dF0x = np.gradient(F0, h, h)
dF2y, dF2x = np.gradient(F2, h, h)

# Set figure size for the first set of 3D plots
rcParams['figure.figsize'] = [16, 8]

# Create subplots for two 3D plots side by side
fig, axs = plt.subplots(1, 2, subplot_kw={'projection': '3d'})

# Plot the surface and contour plots for F0 on the first subplot
axs[0].plot_surface(X, Y, F0, cmap='gray', linewidth=0, antialiased=False)
axs[0].contour(X, Y, F0, zdir='z', offset=axs[0].get_zlim()[0], cmap='gray')

# Plot the surface and contour plots for F2 on the second subplot
axs[1].plot_surface(X, Y, F2, cmap='gray', linewidth=0, antialiased=False)
axs[1].contour(X, Y, F2, zdir='z', offset=axs[0].get_zlim()[0], cmap='gray')

# Display the first set of 3D plots
plt.show()
```
