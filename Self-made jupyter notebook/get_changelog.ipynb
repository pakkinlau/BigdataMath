```python
import os
import datetime

def get_changelog(root_folder):
    changelog = []
    for foldername, subfolders, filenames in os.walk(root_folder):
        for filename in filenames:
            if filename.endswith('.md'):
                filepath = os.path.join(foldername, filename)
                last_modified_timestamp = os.path.getmtime(filepath)
                last_modified_date = datetime.datetime.fromtimestamp(last_modified_timestamp)
                formatted_date = last_modified_date.strftime('%d-%m-%Y %H:%M')
                changelog.append((filename, formatted_date))

    # Sort changelog entries by last modified timestamp in descending order
    changelog.sort(key=lambda x: x[1], reverse=True)

    return changelog

def print_changelog(changelog):
    print("file         | last updated")
    print("--------------------------")
    for filename, last_updated in changelog:
        print(f"{filename.ljust(12)} {last_updated}")

# Example usage
if __name__ == "__main__":
    changelog = get_changelog(root_folder)
    print_changelog(changelog)
```
